```python
import numpy as np
from math import *
import matplotlib.pyplot as plt
```

$G(x) = x- f(x) = 0$; pero $f(x)=x³ +7$

$G(x) = x-x³-7=0$

```python
def f(x):
    return x**3 +7
def G(x):
    return x-f(x)
```

```python
x1=-10
x2=10
xp=(x1+x2)/2

error=1.0
umbral=1e-6
while error > umbral:
    
    if G(x1)/abs(G(x1)) == G(xp)/abs(G(xp)):
        xp_ant= xp 
        x1=xp
        xp =(x1+x2)/2
    elif G(x2)/abs(G(x2)) == G(xp)/abs(G(xp)):
        xp_ant= xp 
        x2=xp
        xp =(x1+x2)/2
    error=abs(G(xp) - G(xp_ant))
    
print(xp)
```
